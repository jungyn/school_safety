# 파일불러오기

```python
data1_2020 = pd.read_excel("data_2020.xlsx") #2020년 데이터 불러오기
data1_2021 = pd.read_excel("data_2020.xlsx") #2021년 데이터 불러오기
data1_2022 = pd.read_excel("data_2020.xlsx") #2022년 데이터 불러오기
```

# 시각화 - 2020

```python
import matplotlib

matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False
```

## 지역별

```python
fig=plt.figure(figsize=(10,10), dpi=100)
ax=fig.subplots()

#카운트 그래프 사용
sns.countplot(y = '지역', data = data1_2020, order = data1_2020['지역'].value_counts().index, palette = "Pastel1")
#수가 제일 많은 순서대로

_=ax.set_title('지역',fontsize=15) #제목 붙이기

#사고수 넘버링
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', (p.get_width() + 0.5, p.get_y() + 0.5), ha='center', va = 'center', rotation = -90, fontsize=8)
   
plt.xlabel("사고 수") #x축 라벨

#그래프 그리기
plt.show()
```

경기도 사고수가 7259명, 서울 사고수가 4271명으로 제일 많고 세종시 사고수가 628명으로 제일 적음

## 학교급

```python
counts = data1_2020['new_학교급'].value_counts()
labels = ['중', '초', '고', '유','기타']

colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0', "lightgray"]
# 파이 차트
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, colors = colors)

# 그래프 제목 추가
plt.title('학교급별 사고수')

# 그래프 표시
plt.show()
```

중학생의 사고비율이 28.6%, 초등학생의 사고비율이 27.9%, 고등학생의 사고비율이 27.4% 유치원생의 사고비율이 15.0%, 기타 학급의 사고비율이 1.1%를 차지함
